```python
import pandas as pd
```

```python
data = pd.read_excel('daily_means with variables.xlsx')
```

```python
data_Air_Temp = data[['Date', 'Air Temp']]  
data_Air_Temp.set_index('Date', inplace=True)
data_Air_Temp.head()
```

```python
data_Air_Temp.info()
```

```python

data_Air_Temp = data_Air_Temp.sort_index()


start_date = data_Air_Temp.index.min()
end_date = data_Air_Temp.index.max()

date_range = pd.date_range(start=start_date, end=end_date)

missing_dates = date_range.difference(data_Air_Temp.index)
num_missing_days = len(missing_dates)

print(f"Number of Missing Days: {num_missing_days}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(data_Air_Temp['Air Temp'], color='grey')

plt.xlabel('Date')
plt.ylabel('Tempreture')
plt.title('Air Tempreture')

plt.show()
```
